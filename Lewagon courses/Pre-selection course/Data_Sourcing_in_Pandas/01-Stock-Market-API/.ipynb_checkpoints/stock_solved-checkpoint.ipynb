# Stock prices

```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib
import requests
```

```python
def fix_datetime_on_dataframe(df):
    df['date'] = pd.to_datetime(df['date'])
    df.set_index(keys='date', inplace=True)
```

```python
def create_stock_df_of_company(company_code):
    url = f'http://iex.lewagon.com/stable/stock/{company_code}/chart/3m'
    df = pd.read_json(url)
    fix_datetime_on_dataframe(df)
    df['company_code'] = company_code
    return df
```

```python
apple_df = create_stock_df_of_company('aapl')
apple_df.head()
```

```python
apple_df['close'].plot(kind='line')
```

```python
apple_df[['open', 'close', 'high', 'low']].plot(kind='line', figsize=(10,10))
```

```python
codes = ['amzn', 'googl', 'fb']
stock_dfs = [apple_df]
for code in codes:
    stock_dfs.append(create_stock_df_of_company(code))
```

```python
stocks_df = pd.concat(stock_dfs)
stocks_df.head()
# stocks_df_ = stocks_df.loc[:,['uClose','company_code']]
# stocks_df_
```

```python
def normalize_pivot(pivot_df):
    new_df = pivot_df.copy()
    for column in new_df:
        #new_df[column]=new_df[column]-new_df[column].mean()/new_df[column].std()
        new_df[column]=(new_df[column]-new_df[column].min())/(new_df[column].max()-new_df[column].min())
    return new_df
```

```python
# get ingations solution
pivot_df = stocks_df.pivot(columns='company_code', values='close')
pivot_df.plot(kind='line', figsize=(8,8))
```

```python
normalized_pivot = normalize_pivot(pivot_df)
normalized_pivot.plot(kind='line', figsize=(8,8))
```
